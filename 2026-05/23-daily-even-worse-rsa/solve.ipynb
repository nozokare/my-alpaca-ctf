```sage
with open("handout/output.txt") as f:
    data = dict([line.strip().split(" = ") for line in f.readlines()])

p = int(data["p"])
e = int(data["e"])
c = int(data["c"])
```

```sage
g, d, k = xgcd(e, p - 1) # e*d + (p-1)*k = g
mg = pow(c, d, p) # (m^e)^d = m^g * m^(p-1)^{-k} ≡ m^g (mod p)
```

```sage
R.<x> = PolynomialRing(Zmod(p))
f = x^6 - mg
for m in f.roots():
    print(bytes.fromhex(hex(m[0])[2:]))
```
