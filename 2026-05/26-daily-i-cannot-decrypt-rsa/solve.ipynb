```python
import ast
from Crypto.Util.number import bytes_to_long, long_to_bytes

with open("handout/output.txt") as f:
    data = dict([line.strip().split(" = ") for line in f.readlines()])

n = int(data["n"])
e = int(data["e"])
phi = int(data["phi"])
flag = ast.literal_eval(data["flag"])
```

```python
import math
a = 1
b = -(phi - n - 1)
c = n
D = b**2 - 4*a*c
assert D == math.isqrt(D) ** 2

p = (-b + math.isqrt(D))//2
q = (-b - math.isqrt(D))//2
assert p * q == n
```

```python
d = pow(e, -1, phi)
delta = pow(e * d, -1, (p - 1) * (q - 1))
m_ = bytes_to_long(flag)
m = pow(m_, delta, n)
long_to_bytes(m)
```
