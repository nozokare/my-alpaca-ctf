```python
n = A.ncols()
m = A.nrows()
B = transpose(block_matrix(ZZ, [[A, p*identity_matrix(m)]]))
print(f"n={n}, m={m}, B.shape={B.nrows()}x{B.ncols()}")
B
```

```python
BB = B.LLL()[n:]
BB
```

```python
def babai_cvp(B, w):
    n = B.nrows()
    BB, _ = B.gram_schmidt()
    e = w
    for i in range(n)[::-1]:
        c = (e.dot_product(BB[i]) / BB[i].dot_product(BB[i])).round()
        e -= c * B[i]
    return w - e

v = babai_cvp(BB, b.change_ring(ZZ))
v
```

```python
s = v[:n] * A[:n].transpose() ^ -1
s
```
