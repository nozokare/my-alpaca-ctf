```python
with open("handout/output.txt") as f:
    s1 = int(f.readline().split(":")[1])
    s2 = int(f.readline().split(":")[1])
    s3 = int(f.readline().split(":")[1])
    s4 = int(f.readline().split(":")[1])
    flag_enc = bytes.fromhex(f.readline().split(":")[1])
```

```python
d2 = s2 - s1
d3 = s3 - s2
d4 = s4 - s3

d2 * d4 - d3 * d3
```

```python
from sympy import factorint
factorint(d3 * d3 - d2 * d4)
```

```python
p = 13076220846716751461
a = d3 * pow(d2, -1, p) % p
b = (s2 - a * s1) % p
s0 = (s1 - b) * pow(a, -1, p) % p
print(f"{p=}, {a=}, {b=}, {s0=}")
```

```python
import hashlib
from Crypto.Cipher import AES
from Crypto.Util.Padding import unpad

def decrypt_flag(ct, params):
    key = hashlib.sha256(str(params).encode()).digest()
    cipher = AES.new(key, AES.MODE_CBC, iv=ct[:16])
    return unpad(cipher.decrypt(ct[16:]), 16)

decrypt_flag(flag_enc, (p, a, b, s0))
```
