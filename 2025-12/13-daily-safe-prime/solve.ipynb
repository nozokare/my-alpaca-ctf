```python
with open("handout/output.txt") as f:
    data = dict([line.strip().split(" = ") for line in f.readlines()])

n = int(data["n"])
e = 65537
c = int(data["c"])
```

```python
import math

a = 2
b = 1
c = -n

D = b * b - 4 * a * c
assert math.isqrt(D) ** 2 == D

p = (-b + math.isqrt(D)) // (2 * a)
q = 2 * p + 1
assert p * q == n
```
