```python
from pwn import remote
```

```python
host, port = "localhost", 9999
```

```python
with remote(host, port) as conn:
    conn.sendlineafter(b"FLAG: ", b"Alpaca{")
    conn.recv(1024)
```

```python
with remote(host, port) as conn:
    conn.sendlineafter(b"FLAG: ", b"Alpaca*")
    conn.recv(1024)
```
